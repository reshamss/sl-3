```python
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
# Select numeric columns only
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Setup the subplot grid dynamically
n_cols = 2
n_rows = math.ceil(len(numeric_columns) / n_cols)

# Set Seaborn style
sns.set(style="whitegrid")

# Create figure
plt.figure(figsize=(n_cols * 6, n_rows * 4))

# Plot each boxplot
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.boxplot(y=data[column], color='skyblue')
    plt.title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()
```

```python
# Filter dataset for 'Iris-setosa', 'Iris-versicolor', and 'Iris-virginica'
setosa_df = df[df['Species'] == 'Iris-setosa']
versicolor_df = df[df['Species'] == 'Iris-versicolor']
virginica_df = df[df['Species'] == 'Iris-virginica']
```

```python
# Plot boxplots for all features to detect outliers
plt.figure(figsize=(12, 8))

# Create a boxplot for each numeric feature
sns.boxplot(data=df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])

# Set plot title
plt.title('Boxplot for Iris Features to Detect Outliers')
plt.show()
```

```python
# Scatter plot for SepalLengthCm vs SepalWidthCm
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['SepalLengthCm'], y=df['SepalWidthCm'], hue=df['Species'], palette='Set1')

# Show the plot
plt.title('Sepal Length vs Sepal Width (with Outliers)')
plt.show()
```

```python
# Histogram for SepalLengthCm
plt.figure(figsize=(8, 6))
sns.histplot(df['SepalLengthCm'], bins=15, kde=True, color='lightblue')

# Add lines for outlier ranges (optional, for visualization)
plt.axvline(x=df['SepalLengthCm'].mean() + 3*df['SepalLengthCm'].std(), color='red', linestyle='dashed', linewidth=2, label='Outlier Threshold (+3 Std Dev)')
plt.axvline(x=df['SepalLengthCm'].mean() - 3*df['SepalLengthCm'].std(), color='red', linestyle='dashed', linewidth=2, label='Outlier Threshold (-3 Std Dev)')

# Show plot with title and labels
plt.title('Histogram for Sepal Length (with Outlier Threshold)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Countplot (Barplot) to show number of samples per species
plt.figure(figsize=(8, 6))
sns.countplot(x='Species', data=df, palette='Set1')

# Set plot title
plt.title('Count of Each Species in the Iris Dataset')
plt.show()
```

```python
# Count the occurrences of each species
species_counts = df['Species'].value_counts()

# Create a pie chart for species distribution
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral'])

# Set the title
plt.title('Distribution of Iris Species')

# Display the pie chart
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Initialize the scaler
scaler = MinMaxScaler()

# Apply Min-Max scaling to the numeric features
scaled_data = scaler.fit_transform(df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])

# Convert the scaled data back to a DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])

# Show the scaled data
print(scaled_df.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Apply standard scaling to the numeric features
scaled_data = scaler.fit_transform(df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])

# Convert the scaled data back to a DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])

# Show the scaled data
print(scaled_df.head())
```
